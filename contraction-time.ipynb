```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
mass = np.array([60, 25, 15, 9, 5, 3, 2, 1.5, 1, 0.8])
contraction_time  = np.array([0.0282, 0.0708, 0.117, 0.288, 1.15, 7.24, 23.4, 35.4, 38.9, 68.4])
```

```python
fig, ax = plt.subplots()
ax.semilogx(mass, contraction_time, linewidth=3)
ax.set_xlabel(r"$M/M_\odot$", fontsize=16)
ax.set_ylabel("Temps de contraction (Ma)", fontsize=16)
ax.set_ylim(-3, 70)
fig.savefig("./chap07/contraction-time.png")
```
